```python
def XsecToTs(xsec, num_density_unit_t,t):
    return np.exp(- num_density_unit_t * t * xsec)
```

```python
ar_bottle_pressure = 197.385 * 1e5 # in Pa (SI unit)
ar_bottle_temp = 293.0 # in Kelvin
num_density_unit_t=(ar_bottle_pressure / (8.31446261815324 * ar_bottle_temp * 1e6)) * 6.02214076e23 * 1e-24
```

```python
f = open("temp_data/Ar.txt")
energy = []
xsec = []
for line in f:
    energy.append(float(line[3:14])*1e6) # in eV
    xsec.append(float(line[16:27]))
    #xsec.append(float(line[20:27]))
energy = np.array(energy)
xsec = np.array(xsec)
f.close()
```

```python
##Set t_true and Num of bins
N_bins = 100
en_bins = np.logspace(3,5,N_bins)
Xsec = np.interp(en_bins, energy, xsec)
t_true = 12
ts = XsecToTs(Xsec,num_density_unit_t,t_true)
```

```python
N=100
Target_in = np.array([])
Target_out = np.array([])
for i in range(N_bins):
    Target_out = np.append(Target_out,np.random.poisson(N))
    Target_in = np.append(Target_in,np.random.poisson(N*ts[i]))
T_list = Target_in/Target_out    
```

```python
plt.hist(T_list)
```

```python
def p(T,t,N,xsec):
    T_true = XsecToTs(xsec,num_density_unit_t,t)
    A=(T_true + T**2)/(2*N*T_true)
    C=(1+T_true)*N/2
    B=1+T
    I=(1/(4* A**(3/2)))*np.exp(-C)*(np.sqrt(np.pi)*B*(-np.exp(B**2/4/A))*sc.erfc(B/2/np.sqrt(A))+ 2*np.sqrt(np.pi)*B*(np.exp(B**2/4/A))- np.sqrt(np.pi)*B*(np.exp(B**2/4/A))*float(mpmath.gammainc(-1/2,B**2/4/A))+2*np.sqrt(A))
    return I/(2*np.pi*N*np.sqrt(T_true))
```

```python
def logli(t):
    L=0
    N=100
    for i in range(N_bins):
        L=L + np.log(p(T_list[i],t,N,Xsec[i]))
    return -2*L
```

```python
t_fit = np.linspace(8, 18, num=100)
L=np.array([])
for i in t_fit:
    L=np.append(L,logli(i))
```

```python
plt.plot(t_fit, L,label='t_true = 12')
plt.legend(loc='upper right')
plt.xlabel('t_fit')
plt.ylabel('-2lnL')
```

```python
m_L = Minuit(logli,12)
m_L.migrad()  
```

```python
t_fit = []
t_fit_err = []
N=100 ##num of events
N_bins = 100 ##num of bins
en_bins = np.logspace(3,5,N_bins)
Xsec = np.interp(en_bins, energy, xsec)
t_true = 11 ##fit parameter
ts = XsecToTs(Xsec,num_density_unit_t,t_true)##T_true for each bin
for i in range(2000):
    Target_in = np.array([])
    Target_out = np.array([])
    for i in range(N_bins):
        Target_out = np.append(Target_out,np.random.poisson(N))
        Target_in = np.append(Target_in,np.random.poisson(N*ts[i]))
    T_list = Target_in/Target_out 
    def logli(t):
        L=0
        N=100
        for i in range(N_bins):
            L=L + np.log(p(T_list[i],t,N,Xsec[i]))
        return -2*L
    m_L = Minuit(logli,11)
    m_L.migrad()  
    t_fit.append(m_L.values["t"])
    t_fit_err.append(m_L.errors["t"])
t_fit = np.array(t_fit)
t_fit_err = np.array(t_fit_err)      
```

```python
plt.hist((t_fit - t_true)/t_fit_err,bins=25,label = 't_true = 11')
plt.xlabel('(t_fit - t_true) / sigma_fit')
plt.legend(loc='upper right')
plt.savefig('t_true = 11')
```
